```python
import time
import cv2
import numpy as np

from ultralytics import YOLO

from config import YOLO_MODELS


class PlateDetector:

    def __init__(self):

        self.models = {}

    def load_model(self, model_name):

        if model_name not in self.models:

            self.models[model_name] = YOLO(
                str(YOLO_MODELS[model_name])
            )

        return self.models[model_name]

    def detect(self, image, model_name):

        model = self.load_model(model_name)

        start = time.time()

        results = model.predict(

            source=image,

            conf=0.25,

            verbose=False

        )

        inference_time = time.time() - start

        result = results[0]

        img = result.plot()

        boxes = []

        if result.boxes is not None:

            for box in result.boxes:

                x1, y1, x2, y2 = map(
                    int,
                    box.xyxy[0]
                )

                conf = float(box.conf[0])

                boxes.append({

                    "bbox": (x1, y1, x2, y2),

                    "confidence": conf

                })

        return {

            "image": img,

            "boxes": boxes,

            "time": inference_time

        }
```

```python
import easyocr
import cv2

from config import OCR_LANG


class PlateOCR:

    def __init__(self):

        self.reader = easyocr.Reader(

            OCR_LANG,

            gpu=True

        )

    def recognize(

        self,

        image,

        bbox

    ):

        x1, y1, x2, y2 = bbox

        plate = image[

            y1:y2,

            x1:x2

        ]

        result = self.reader.readtext(

            plate,

            detail=1

        )

        if len(result) == 0:

            return "", 0

        text = result[0][1]

        confidence = float(result[0][2])

        return text, confidence
```

```python
import pandas as pd

models = {
    "YOLOv5":"results_yolov5.csv",
    "YOLOv8":"results_yolov8.csv",
    "YOLOv10":"results_yolov10.csv"
}

table=[]

for model,file in models.items():

    df=pd.read_csv(file)

    last=df.iloc[-1]

    table.append({

        "Model":model,

        "Precision":last["metrics/precision(B)"],

        "Recall":last["metrics/recall(B)"],

        "mAP50":last["metrics/mAP50(B)"],

        "mAP50-95":last["metrics/mAP50-95(B)"]

    })

comparison=pd.DataFrame(table)

comparison
```
